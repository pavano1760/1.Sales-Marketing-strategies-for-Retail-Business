### 3.Problem Statement : To check the Price Elasticity of demand (PED) of Webcams. 
###    Data: monthwise four years data of Quantity and Price of webcam product.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from statsmodels.compat import lzip
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
df=pd.read_excel('C:/datalake/project_retail/master_dataset.xlsx', sheet_name='price_data',encoding = 'unicode_escape')
df.head(3)
```

### Plotting Quality vs Price Data to understand relation of data.

```python
sns.set_style('whitegrid')
ax = sns.lineplot(x=df['Quantity'], y=df['Price'],data=df)
plt.title('Price vs Quantity')
plt.figure(figsize=(10,18))
plt.show()
```

### Price is Inversaly proportional to the Quantity.

### Now let us find Elasticity type? ...... it is Inelastic or Unitary elastic or Elastic???

#### applying statastical model for detail result summary.

```python
elst_model = ols("Quantity ~ Price", data=df).fit()
print(elst_model.summary())
```

### We concluding from this result summary:-

#### 1.P-value is -0.0308  It is less than one (p-value  <  1).so this is Inelastic Elasticity. 
#### 2.PED comes out to be -0.0308. It shows that when the price of Webcam is increased by 1% the demand for beef falls by 0.53%
#### 3.The small P values indicate that we can reject the null hypothesis that Price has no effect on Quantity.

### Plotting the result by Regression plots.

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(elst_model, fig=fig)
```

```python
# Regression plots
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(elst_model, 'Price', fig=fig)
```
